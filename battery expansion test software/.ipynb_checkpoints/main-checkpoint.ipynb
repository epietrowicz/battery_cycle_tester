```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
batt_data = pd.read_csv('log.txt')
batt_data.columns = ['time', 'volts', 'current', 'state']
batt_data.head()
```

```python
batt_data.plot(kind='line', x='time', y='volts', color='red')
plt.show()
```

```python
#state = 0 discharging
#state = 1 charging
batt_data.plot(kind='line', x='time', y='state', color='blue')
plt.show()
```

```python
batt_data.plot(kind='line', y='time', use_index=True, color='green')
plt.show()
```

```python
start_test_time = batt_data.loc[0, 'time']
end_test_time = batt_data.loc[len(batt_data)-1, 'time']
test_duration = end_test_time - start_test_time
test_duration = ((test_duration / 1000) / 3600)
print("Duration of the test (hrs):")
print(test_duration)
```

```python
avg_current = 0
for i in range(len(batt_data)):
    avg_current = avg_current + batt_data.loc[i, 'current']
avg_current = avg_current / len(batt_data)
avg_current = avg_current * 1000
print("Average Current Draw (mA): ")
print(avg_current)
```

```python
start_flag = True
end_flag = True

for i in range(len(batt_data)):
    if batt_data.loc[i,'state'] == 0 and start_flag:
        start_time = (batt_data.loc[i, 'time'])
        start_flag = False
    try:
        if batt_data.loc[i+1, 'state'] == 1 and end_flag:
            end_time = (batt_data.loc[i, 'time'])
            end_flag = False
    except:
        end_time = (batt_data.loc[i, 'time'])
        end_flag = False

discharge_time = ((end_time - start_time) / 1000) / 3600

print("Time to discharge: (hrs)")
print(discharge_time)
```

```python
capacity = avg_current * discharge_time
print("Calculated capacity: (mAh)")
print(capacity)
```
